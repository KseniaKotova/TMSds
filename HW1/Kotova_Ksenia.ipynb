```python
df = df.drop(['Alley','PoolQC','Fence','MiscFeature'],axis=1)
```

```python
object_columns_df = df.select_dtypes(include=['object'])
numerical_columns_df =df.select_dtypes(exclude=['object'])
```

```python
object_columns_df.dtypes
```

```python
null_counts = object_columns_df.isnull().sum()
print(null_counts)
```

```python
columns_None = ['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','GarageType','GarageFinish','GarageQual','FireplaceQu','GarageCond']
object_columns_df[columns_None]= object_columns_df[columns_None].fillna('None') #заполнение None
```

```python
columns_with_lowNA = ['MasVnrType','Electrical']
#заполнение модой
object_columns_df[columns_with_lowNA] = object_columns_df[columns_with_lowNA].fillna(object_columns_df.mode().iloc[0])
```

```python
null_counts = numerical_columns_df.isnull().sum()
print(null_counts)
```

```python
columns_with_lowNA = ['LotFrontage', 'GarageYrBlt', 'MasVnrArea']

#заполнение средним
for col in columns_with_lowNA:
    numerical_columns_df[col] = numerical_columns_df[col].fillna(numerical_columns_df[col].mean())
```

```python
df_final = pd.concat([object_columns_df, numerical_columns_df], axis=1,sort=False)
df_final.head()
```

```python
sns.histplot(df_final["SalePrice"], color="blue")
plt.show()
```

```python
price_transformed = np.log(df_final['SalePrice']) #трансформация с помощью логарифмирования с целью приближения к нормальному распределению
sns.histplot(price_transformed, color='blue')
plt.show()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df_final.corr(), vmax=1, vmin=-1, annot=True)
```
