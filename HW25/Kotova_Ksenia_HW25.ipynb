#### Найдены ошибки в названиях машин: maxda = mazda, Nissan = nissan, porsche = porcshce, toyota = toyouta, vokswagen = volkswagen = vw

```python
def correct_brand_name(car_name):
    brand_model = car_name.split(' ', 1)
    brand = brand_model[0].lower()
    model = brand_model[1] if len(brand_model) > 1 else ''

    brand_correction = {
        'maxda': 'mazda',
        'nissan': 'nissan',
        'porcshce': 'porsche',
        'toyouta': 'toyota',
        'vokswagen': 'volkswagen',
        'vw': 'volkswagen'
    }
    corrected_brand = brand_correction.get(brand, brand)
    return f'{corrected_brand} {model}'


data['CarName'] = data['CarName'].apply(correct_brand_name)
```

```python
def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df


cars_lr = dummies('fueltype',cars_lr)
cars_lr = dummies('aspiration',cars_lr)
cars_lr = dummies('carbody',cars_lr)
cars_lr = dummies('drivewheel',cars_lr)
cars_lr = dummies('enginetype',cars_lr)
cars_lr = dummies('cylindernumber',cars_lr)
cars_lr = dummies('datarange',cars_lr)
```

## Tensorflow

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
import tensorflow as tf

X_train_tensor = tf.convert_to_tensor(X_train.values, dtype=tf.float32)
y_train_tensor = tf.convert_to_tensor(y_train.values, dtype=tf.float32)

num_features = X_train_tensor.shape[1]


w = tf.Variable(tf.random.normal([num_features, 1]))
b = tf.Variable(tf.zeros([1]))

learning_rate = 0.01

def compute_loss():
    y_pred = tf.matmul(X_train_tensor, w) + b
    loss = tf.reduce_mean((y_train_tensor - y_pred) ** 2)
    return loss

loss_values = []


for step in range(1000):
    with tf.GradientTape() as tape:
        loss = compute_loss()
    gradients = tape.gradient(loss, [w, b])
    w.assign_sub(learning_rate * gradients[0])
    b.assign_sub(learning_rate * gradients[1])

    loss_values.append(loss.numpy())
    
    if step % 100 == 0:
        print(f"Step: {step}, Loss: {loss.numpy()}")


plt.plot(loss_values)
plt.title('Loss values')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.show()

print(f"w = {w.numpy()}, b = {b.numpy()}")
```

```python
X_test_tensor = tf.convert_to_tensor(df_test.drop('price', axis=1).values, dtype=tf.float32)
y_test_tensor = tf.convert_to_tensor(df_test['price'].values, dtype=tf.float32)

y_pred_tensor = tf.matmul(X_test_tensor, w) + b
test_loss = tf.reduce_mean((y_test_tensor - y_pred_tensor) ** 2)

print(f"Test Loss: {test_loss.numpy()}")
```

## Pytorch

```python
import torch
import matplotlib.pyplot as plt


X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1) 

num_features = X_train_tensor.shape[1]
w = torch.randn(num_features, 1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
learning_rate = 0.01

def compute_loss():
    y_pred = torch.matmul(X_train_tensor, w) + b
    loss = torch.mean((y_train_tensor - y_pred) ** 2)
    return loss

loss_values = []

for step in range(1000):
    loss = compute_loss()
    loss.backward()

    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad
        w.grad.zero_()
        b.grad.zero_()

    loss_values.append(loss.item()) 

    if step % 100 == 0:
        print(f"Step: {step}, Loss: {loss.item()}")

# Получаем значения весов и смещения
w_value = w.detach().numpy()
b_value = b.item()

print(f"w = {w_value}, b = {b_value}")

# График ошибки по итерациям
plt.plot(loss_values)
plt.title('Loss values')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.show()
```

```python
X_test_tensor = torch.tensor(df_test.drop('price', axis=1).values, dtype=torch.float32)
y_test_tensor = torch.tensor(df_test['price'].values, dtype=torch.float32).view(-1, 1)

y_pred_tensor = torch.matmul(X_test_tensor, w) + b
test_loss = torch.mean((y_test_tensor - y_pred_tensor) ** 2)

print(f"Test Loss: {test_loss.item()}")
```

## В результате меньшее значение ошибки на тестовой выборке получилось при использовании pytorch - 0.10895461589097977. А при использовании tensorflow - 0.26578184962272644.

# <strong>MNIST</strong>

```python
train_df = pd.read_csv('fashion-mnist_train.csv',sep=',')
test_df = pd.read_csv('fashion-mnist_test.csv', sep = ',')
```

```python
train_data = np.array(train_df, dtype = 'float32')
test_data = np.array(test_df, dtype='float32')

x_train = train_data[:,1:]/255
y_train = train_data[:,0]
x_test= test_data[:,1:]/255
y_test=test_data[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_train, y_train, test_size = 0.2, random_state = 42)
```

```python
class_names = ['T_shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize=(10, 10))
for i in range(36):
    plt.subplot(6, 6, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i].reshape((28,28)))
    label_index = int(y_train[i])
    plt.title(class_names[label_index])
plt.show()
```

```python
from tensorflow.keras.callbacks import TensorBoard
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers, losses
```

```python
tb_callbacks = TensorBoard(log_dir='log1', histogram_freq=1)

image_rows = 28
image_cols = 28
batch_size = 4096
image_shape = (image_rows,image_cols,1)

optimizer=optimizers.Adam(learning_rate=0.01)
```

```python
X_train = X_train.reshape(X_train.shape[0],*image_shape)
X_test = X_test.reshape(X_test.shape[0],*image_shape)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(image_rows, image_cols, 1)),
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])

model.compile(optimizer=optimizer, loss=losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5, batch_size=batch_size, callbacks=[tb_callbacks], validation_data=(X_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Accuracy on the test set: {test_acc * 100:.2f}%")
```

## При tf.keras.layers.Dense(64, activation="relu"), tf.keras.layers.Dense(10, activation="softmax"), optimizer=optimizers.Adam(learning_rate=0.01), 5 эпох:

```python
test_loss, test_acc = model_1.evaluate(X_test, y_test)
print(f"Accuracy on the test set: {test_acc * 100:.2f}%")
```

## При tf.keras.layers.Dense(32, activation="relu"), epochs=10


## По сравнению с предыдущим вариантом ошибка уменьшилась и точность выросла. Похоже на то, что с такой архитектурой это максимальная точность, которую можно достигнуть

```python
tb_callbacks = TensorBoard(log_dir='log3', histogram_freq=1)

model_2 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(image_rows, image_cols, 1)),
    tf.keras.layers.Dense(500, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])

model_2.compile(optimizer=optimizer, loss=losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
model_2.fit(X_train, y_train, epochs=10, batch_size=batch_size, callbacks=[tb_callbacks], validation_data=(X_test, y_test))
```

```python
test_loss, test_acc = model_2.evaluate(X_test, y_test)
print(f"Accuracy on the test set: {test_acc * 100:.2f}%")
```

## При tf.keras.layers.Dense(500, activation="relu") точность слегка увеличилась.


## Сверточная нейронная сеть

```python
tb_callbacks = TensorBoard(log_dir='log4', histogram_freq=1)

cnn_model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape = (image_rows, image_rows, 1)),
    tf.keras.layers.MaxPooling2D(pool_size=2),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(10,activation = 'softmax')
])

cnn_model.compile(optimizer=optimizer, loss=losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
cnn_model.fit(X_train, y_train, epochs=10, batch_size=batch_size, callbacks=[tb_callbacks], validation_data=(X_test, y_test))

test_loss, test_acc = cnn_model.evaluate(X_test, y_test)
print(f"Accuracy on the test set: {test_acc * 100:.2f}%")
```

## Результат обучения сверточной нейронной сети с данной архитектурой очень схож с результатом предыдущей архитектуры. Точность получилась немного выше у сверточной сети.
